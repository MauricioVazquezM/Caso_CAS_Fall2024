```python
# Partiendo el dataset original en cuantiles, ¿como se ve la evolucion de las sumas aseguradas?
df_aux.sort_values(by='suma_asegurada', ascending=False, inplace=True)
df_sa = df_aux[df_aux['suma_asegurada'] > 0]
df_sa['suma_asegurada'] = df_sa['suma_asegurada']/1000000
df_sa['cuantiles_suma_asegurada'] = pd.qcut(df_sa['suma_asegurada'], q=4)
df_sa['cuantiles_suma_asegurada'] = df_sa['cuantiles_suma_asegurada'].apply(lambda x: f"({float(x.left):.3f}, {float(x.right):.3f}]")
#df_sa['cuantiles_suma_asegurada'] = df_sa['cuantiles_suma_asegurada'].astype(str)

# Crear tabla de frecuencia resumida (crosstab)
heatmap_data = pd.crosstab(df_sa['cuantiles_suma_asegurada'], df_sa['año_póliza'])

## Definir los valores únicos de los años en la columna x
unique_years = sorted(df_sa['año_póliza'].unique())
unique_cuantiles = ['(4.998, 54.614]', '(54.614, 123.296]', '(123.296, 289.950]', '(289.950, 1820.999]']

# Crear el heatmap
heatmap = go.Heatmap(
    z=heatmap_data.values,  
    x=heatmap_data.columns,  
    y=heatmap_data.index,    
    colorscale='blues',     
    text=heatmap_data.values,  
    texttemplate="%{text}",   
    textfont={"size":12},     
    showscale=True,           
)

# Crear la figura
fig = go.Figure(data=[heatmap])

# Añadir título y etiquetas
fig.update_layout(
    title='Conteo por cuantiles de sumas aseguradas <br> (en millones)',
    template='plotly_white',
    xaxis=dict(
        tickmode='array',       
        tickvals=unique_years,   
        ticktext=unique_years,          
        title='Año poliza'            
    ),
    yaxis=dict(
        tickmode='array',       
        tickvals=unique_cuantiles,   
        ticktext=unique_cuantiles,          
        title='Conteo por cuantiles de suma asegurada <br> (en millones)'            
    )
)

# Mostrar la figura
fig.show()
```

```python
# Evolucion de la tasa pura de riesgo: TPR = Suma del Coste de Siniestros / Suma Asegurada Total
df_grouped = df_sa.groupby('año_póliza')[['monto_de_siniestro', 'suma_asegurada']].sum().reset_index()
df_grouped.columns = ['Año Póliza', 'Suma MS', 'Suma SA']
df_grouped['tpr_por_año'] = df_grouped['Suma MS'] / df_grouped['Suma SA']

mean = df_grouped['tpr_por_año'].mean()


fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_grouped['Año Póliza'], 
    y=df_grouped['tpr_por_año'],
    mode='lines+markers',
    name='TPR por Año',
    line=dict(color='blue'),
    marker=dict(size=8)
))
fig.add_trace(go.Scatter(
    x=df_grouped['Año Póliza'], 
    y=[mean] * len(df_grouped['Año Póliza']), 
    mode='lines',
    name='Media TPR',
    line=dict(color='red', dash='dash')
))

# Actualizar el layout del gráfico
fig.update_layout(
    title='Evolución de la Tasa Pura de Riesgo (TPR) <br>(por Año)',
    xaxis=dict(
        tickmode='array',       
        tickvals=unique_years,   
        ticktext=unique_years,          
        title='Año poliza'            
    ),
    yaxis_title='TPR',
    template='plotly_white',
    legend=dict(
        y=1,
        x=1,
        xanchor='right',
        yanchor='top'
    )
)

# Mostrar el gráfico
fig.show()
```

```python
# Evolucion de Índice de siniestralidad = Suma del Coste de Siniestros / Primas 
df_grouped_2 = df_sa.groupby('año_póliza').agg(
    suma_monto_siniestro=('suma_asegurada', 'sum'),
    conteo_primas=('prima', 'count')).reset_index()
df_grouped_2['indice_siniestralidad_por_año'] = df_grouped_2['suma_monto_siniestro'] / df_grouped_2['conteo_primas']

mean = df_grouped_2['indice_siniestralidad_por_año'].mean()


fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_grouped_2['año_póliza'], 
    y=df_grouped_2['indice_siniestralidad_por_año'],
    mode='lines+markers',
    name='Indice de siniestralidad por año',
    line=dict(color='blue'),
    marker=dict(size=8)
))
fig.add_trace(go.Scatter(
    x=df_grouped_2['año_póliza'], 
    y=[mean] * len(df_grouped_2['año_póliza']), 
    mode='lines',
    name='Media Indice de siniestralidad',
    line=dict(color='red', dash='dash')
))

# Actualizar el layout del gráfico
fig.update_layout(
    title='Evolución del Indice de siniestralidad <br>(por Año)',
    xaxis=dict(
        tickmode='array',       
        tickvals=unique_years,   
        ticktext=unique_years,          
        title='Año poliza'            
    ),
    yaxis_title='Indice de siniestralidad',
    template='plotly_white',
    legend=dict(
        y=1,
        x=0,
        xanchor='left',
        yanchor='top'
    )
)

# Mostrar el gráfico
fig.show()
```

***ANOTACIONES:***

* Existe una relación positiva entre la suma asegurada y la prima, lo que significa que a medida que la suma asegurada aumenta, la prima también tiende a aumentar.
* Sin embargo, la relación no es perfectamente lineal, ya que hay dispersión en los datos, especialmente en las sumas aseguradas más grandes, lo que indica variabilidad en las primas para un mismo valor de suma asegurada.
* Para sumas aseguradas menores a 100 millones, los puntos están muy concentrados cerca de la línea de regresión, lo que sugiere una relación más clara entre ambas variables en este rango.
* A partir de los 100 millones y hasta los 700 millones, los puntos comienzan a dispersarse más, con primas que varían significativamente para la misma suma asegurada. Esto indica que la prima no está tan estrechamente vinculada a la suma asegurada en estos rangos, y que hay otros factores que probablemente influyen en el valor de la prima.
* A partir de los 500 millones, la dispersión es más evidente, con algunos puntos alejados de la línea de regresión, lo que indica una mayor variabilidad en los datos.
* La línea roja representa el ajuste lineal que mejor explica la relación entre la suma asegurada y la prima. Aunque la pendiente es positiva, lo que indica que una mayor suma asegurada generalmente se asocia con una mayor prima, la dispersión de los datos en los valores más altos de suma asegurada sugiere que esta relación es más débil en esos casos.
* El intervalo de confianza del 95% alrededor de la línea de regresión muestra la incertidumbre en el ajuste del modelo. La banda de confianza es relativamente estrecha para valores bajos de suma asegurada, lo que indica que el modelo es más confiable en estos rangos.
* Sin embargo, la banda se ensancha a medida que la suma asegurada aumenta, lo que refleja una mayor incertidumbre sobre la relación entre la prima y la suma asegurada en los rangos altos. Esto sugiere que, para sumas aseguradas grandes, el modelo tiene menos precisión al predecir la prima.
* Se observan algunos outliers o puntos alejados de la tendencia general, especialmente para valores de suma asegurada mayores a 600 millones. Estos puntos podrían indicar casos particulares o atípicos, donde las primas se comportan de manera diferente al resto de los datos.
